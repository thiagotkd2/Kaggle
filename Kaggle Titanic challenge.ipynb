```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Modelling
from sklearn.ensemble import RandomForestClassifier

# read input
test_data = pd.read_csv("./kaggle/input/titanic/test.csv")
train_data = pd.read_csv("./kaggle/input/titanic/train.csv")
```

```python
#encode
age_average_train = train_data.loc[:, 'Age'].mean()
age_average_test = test_data.loc[:, 'Age'].mean()
print(age_average_test)
train_data_encoded = pd.get_dummies(train_data, columns=['Sex'], drop_first=True).fillna(age_average_train)
test_data_encoded = pd.get_dummies(test_data, columns=['Sex'], drop_first=True).fillna(age_average_test)
```

```python
#select and drop
x_train = train_data_encoded.drop(columns=['Survived', 'Name', 'PassengerId', 'Ticket', 'Cabin', 'Embarked'])
x_test = test_data_encoded.drop(columns=['Name', 'PassengerId', 'Ticket', 'Cabin', 'Embarked'])
y_train = train_data_encoded['Survived']
print(test_data_encoded)
```

```python
# now we fit
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
print(y_train)
```

```python
# predict
y_pred = rf.predict(x_test)
output_df = pd.DataFrame()
output_df['PassengerId'] = test_data_encoded['PassengerId']
output_df['Survived'] = y_pred
print(output_df)


output_df.to_csv("./kaggle/working/submission.csv", index=False)
```
